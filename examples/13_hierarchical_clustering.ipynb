# Hierarchical Clustering

Hierarchical clustering builds clusters by progressively merging points.

## Setup: Load the Iris Dataset

```python
import cuanalytics
```

```python
# Load data
df = cuanalytics.load_iris_data()
df.head()
```

## Step 1: Fit Hierarchical Clustering

```python
hier = cuanalytics.fit_hierarchical(
    df,
    formula='~ sepal_length  sepal_width  petal_length  petal_width',
    n_clusters=3,
    linkage='ward'
)
```

## Step 2: Evaluate Clustering

```python
metrics = hier.get_metrics()
metrics
```

## Step 3: Visualize Clusters

```python
hier.visualize()
```
