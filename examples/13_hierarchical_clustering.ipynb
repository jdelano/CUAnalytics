# Hierarchical Clustering

Hierarchical clustering builds clusters by progressively merging points.

## Setup: Load the Iris Dataset

```python
import cuanalytics as ca
```

```python
# Load data
df = ca.load_iris_data()
df.head()
```

## Step 1: Fit Hierarchical Clustering

```python
hier = ca.fit_hierarchical(
    df,
    formula='~ sepal_length + sepal_width + petal_length + petal_width',
    n_clusters=3
)
```

## Step 2: Evaluate Clustering

```python
metrics = hier.get_metrics()
metrics
```

## Step 3: Visualize Results

```python
hier.visualize()  # Dendrogram
hier.visualize_all_features()  # PCA projection using all features
```
