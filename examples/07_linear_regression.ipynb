```python
import cuanalytics as ca

# Load data
df = ca.load_real_estate_data()
# Split data
train, test = ca.split_data(df, test_size=0.2)

# Fit model
model = ca.fit_lm(train, formula='price_per_unit ~ house_age + num_convenience_stores * distance_to_MRT + transaction_date')

# # View summary
model.summary()

# # Visualizations
model.visualize()
# model.visualize_all_features()
# Evaluate

print("\nModel Evaluation:")
test_score = model.score(test)
print(f"Test R²: {test_score['r2']:.4f}")
print(f"Test RMSE: {test_score['rmse']:.4f}")
```
