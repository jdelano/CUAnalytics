```python
from cuanalytics import load_real_estate_data, fit_lm, split_data

# Load data
df = load_real_estate_data()
# Split data
train, test = split_data(df, test_size=0.2)

# Fit model
model = fit_lm(train, formula='price_per_unit ~ house_age + num_convenience_stores * distance_to_MRT + transaction_date')

# View summary
summary = model.summary()

# Visualizations
model.visualize()
model.visualize_all_features()
# Evaluate
train_metrics = model.get_metrics(train)
test_metrics = model.get_metrics(test)

print("\nModel Evaluation:")
print(f"Training R²: {train_metrics['RSquare']:.4f}")
print(f"Test R²: {test_metrics['RSquare']:.4f}")
print(f"\nTraining RMSE: {train_metrics['RMSE']:.4f}")
print(f"Test RMSE: {test_metrics['RMSE']:.4f}")
```
