# Example: Basic Entropy Calculation

## This example shows how to calculate entropy for a simple dataset.

```python
import cuanalytics
```

```python
# Load the mushroom dataset
df = cuanalytics.load_mushroom_data()
```

```python
# Calculate entropy of the target variable
entropy = cuanalytics.calculate_entropy(df['class'])
print(f"Dataset entropy: {entropy:.4f}")
```
