# k-Nearest Neighbors (Regression)

Use KNN to predict a continuous outcome based on nearby points.

## Setup: Load the Real Estate Dataset

```python
import cuanalytics as ca
```

```python
# Load data and create train/test split
df = ca.load_real_estate_data()
train_df, test_df = ca.split_data(df, test_size=0.2, random_state=42)
train_df.head()
```

## Step 1: Fit a KNN Regressor

```python
knn = ca.fit_knn_regressor(train_df, formula='price_per_unit ~ .', k=5)
```

## Step 2: Evaluate Performance

```python
metrics = knn.score(test_df)
print(f"Test RÂ²: {metrics['r2']:.4f}")
print(f"Test RMSE: {metrics['rmse']:.4f}")
```
