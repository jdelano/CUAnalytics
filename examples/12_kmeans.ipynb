# K-Means Clustering

K-Means groups points into k clusters based on distance.

## Setup: Load the Iris Dataset

```python
import cuanalytics as ca
```

```python
# Load data
df = ca.load_iris_data()
df.head()
```

## Step 1: Fit K-Means

```python
kmeans = ca.fit_kmeans(
    df,
    formula='~ sepal_length + sepal_width + petal_length + petal_width',
    n_clusters=2
)
```

## Step 2: Evaluate Clustering

```python
metrics = kmeans.get_metrics()
metrics
```

## Step 3: Visualize Clusters

```python
kmeans.visualize()
```

## Step 4: Describe Clusters with Rules

```python
df_described = kmeans.describe_clusters(max_depth=3)
df_described
```
